```python
import numpy as np
```

```python
a = np.array([0, 100])
```

```python
a = np.array([1, 4, 3, 6 , 7])
b = np.array([100, 56, 89, 78, 23])
```

```python
# mec = marker edge color
# mfc = macker face color
```

```python
plt.plot(a, b, marker = ".", ms = 20, mec = "red", mfc = "yellow")
```

## 

```python
# line styles:
#     1) dashed
#     2) dotted

plt.plot(a, b, marker = ".", ms = 20, mec = "red", mfc = "yellow", linestyle = "dotted")
```

```python
# line width
```

```python
plt.plot(a, b, marker = ".", ms = 20, mec = "red", mfc = "yellow", linestyle = "dashed", linewidth = 3)
```

```python
# line color:
```

```python
plt.plot(a, b, marker = "<", ms = 20, mec = "red", mfc = "yellow", linewidth = 3, color = "pink")
```
