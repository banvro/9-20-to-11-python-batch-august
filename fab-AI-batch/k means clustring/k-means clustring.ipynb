```python
# Elbow method.....
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(1, 12):
    obj = KMeans(n_clusters = i)
    obj.fit_predict(data)
    wcss.append(obj.inertia_)
```

```python
wcss
```

```python
plt.scatter(range(1, 12), wcss, label = "Elbow Points")
plt.plot(range(1, 12), wcss, color = "red", label = "Elbow Curve")
plt.xlabel("Number of Clusters")
plt.ylabel("Calculated WCSS ( Inertia)")
plt.title("Elbow Method")
plt.legend()
plt.grid()
```
