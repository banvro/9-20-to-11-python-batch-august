```python
import numpy as np
```

```python
x = 6 * np.random.rand(200, 1) - 3
y = 0.8 * x**2 + 0.9 * x + 2 + np.random.randn(200, 1)
```

```python
# y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mdl = LinearRegression()
```

```python
mdl.fit(x_train, y_train)
```

```python
y_pred = mdl.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
plt.scatter(x, y)
plt.plot(x_test, mdl.predict(x_test), color = "red")
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 3)
```

```python
x_train_new = poly.fit_transform(x_train)
```

```python
x_train_new
```

```python
x_train
```

```python
x_test_new = poly.transform(x_test)
```

```python
x_test_new
```

```python
new_model = LinearRegression()
```

```python
new_model.fit(x_train_new, y_train)
```

```python
y_pred_new = new_model.predict(x_test_new)
```

```python
r2_score(y_test, y_pred_new)
```

```python
new_model.coef_
```

```python
new_model.intercept_
```

```python
x_new = np.linspace(-3, 3, 200).reshape(200, 1)
x_new_ploy = poly.transform(x_new)
y_new = new_model.predict(x_new_ploy)
```

```python
plt.plot(x_new, y_new, color = "red", linewidth = 4)
plt.scatter(x, y)
```
