```python
zx = np.array([12, 45, 100])
pq = np.array([23, 50, 30])
```

```python
plt.plot(zx, pq)
plt.xlabel("Student Marks")
plt.ylabel("Student Age")
plt.title("Student Information")
```

```python
plt.plot(zx, pq)
plt.xlabel("Student Marks")
plt.ylabel("Student Age")
plt.title("Student Information")
plt.grid()
```

```python
plt.scatter(zx, pq)
plt.xlabel("Student Marks")
plt.ylabel("Student Age")
plt.title("Student Information", loc = "center") # left right center
plt.grid()
```
