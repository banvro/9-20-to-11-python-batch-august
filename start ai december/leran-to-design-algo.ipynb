```python
class myLRAlgo:

    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, x_train, y_train):
        num = 0
        dem = 0

        for i in range(x_train.shape[0]):

            num = num + ((y_train[i] - y_train.mean()) * (x_train[i] - x_train.mean()))
                # 3.51 - 6.3453 * 8.31 - 6.0
            dem = dem + ((x_train[i] - x_train.mean()) * (x_train[i] - x_train.mean()))

            self.m = num / dem
            self.b = y_train.mean() - self.m * x_train.mean()

        print("The solpe of value os m is : ", self.m)
        print("the intersept or value of b is : ", self.b)

    def predict(self, y_test):
        print("i am predicting....")
        y = self.m * y_test - self.b
        print("prediction is : " , y)
```
