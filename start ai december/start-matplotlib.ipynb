```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xyz = np.array([12, 10, 78])
abc = np.array([300, 400, 500])
```

```python
plt.plot(xyz, abc, marker = "o", ms = 20, mfc = "red", mec = "black")
```

```python
plt.plot(xyz, abc, marker = "o", ms = 20, mfc = "red", mec = "black", color = "green", linewidth = 5, linestyle = "dashdot")
```

```python
plt.plot(xyz, abc, marker = "o", ms = 20, mfc = "red", mec = "black", color = "green", linewidth = 5, linestyle = "dashdot")
plt.xlabel("this is x label")
plt.ylabel("this is y lable")
plt.title("this is a title", loc = "right")
plt.grid(axis = "y")
```
