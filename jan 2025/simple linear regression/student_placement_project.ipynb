```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv("/content/updated_placement.csv")
```

```python
data.shape
```

```python
data.info()
```

```python
data.head()
```

```python

plt.scatter(data["cgpa"], data["package"])
plt.xlabel("Cgpa")
plt.ylabel("Package")
plt.title("Student placement record")
```

```python

stu_cgpa = data["cgpa"].values.reshape(-1, 1)
stu_cgpa
```

```python
stu_pkg = data["package"]
stu_pkg
```

```python
x_train, x_test, y_train, y_test = train_test_split(stu_cgpa, stu_pkg, test_size = 0.2)
```

```python
plt.scatter(x_train, y_train, label = "Traing Dataset")
plt.scatter(x_test, y_test, label = "Testing dataset")
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.legend()
```

```python
obj = LinearRegression()
```

```python
obj.fit(x_train, y_train)
```

```python
# Calculate accuracy

x_test
```

```python
y_pred = obj.predict(x_test)
y_pred
```

```python
y_test
```

```python

r2_score(y_test, y_pred)
```

```python
plt.scatter(x_train, y_train, label = "Traing Dataset")
plt.scatter(x_test, y_test, label = "Testing dataset")
plt.plot(x_test, y_pred, label = "Predicted Line", color = "red", linewidth = 4)
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.legend()
```
